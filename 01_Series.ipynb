```python
# Create Series from dict
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = Series(sdata)
obj3
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = Series(sdata, index=states)
obj4
# NaN (missing or not a number) 
```

```python
# Check missing data with isnull, notnull 
pd.isnull(obj4)
```

```python
pd.notnull(obj4)
```
